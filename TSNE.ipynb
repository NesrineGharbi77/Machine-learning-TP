```python
from sklearn.manifold import TSNE
from sklearn.datasets import fetch_openml
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = fetch_openml("mnist_784")
```

```python
x.data.shape
```

```python
data = x.data
```

```python
data[:100][:]
```

```python
tsne = TSNE(n_components=2, perplexity=30, early_exaggeration=12, n_iter=1500)
```

```python
tsne.fit_transform(data[:10000][:])
```

```python
embeddings = [embedding for embedding in tsne.embedding_]
embeddings
```

```python
# prompt: give code that changes array of arrays to pandas dataframe

df = pd.DataFrame(embeddings)
df['Label'] = x.target.astype('string')
```

```python
df.head()
```

```python
df.columns = ['first_axis', 'second_axis', 'Label']
```

```python
df.head()
```

```python
import seaborn as sns
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
x_tranformed = pca.fit_transform(data[:10000][:])

df2 = pd.DataFrame(x_tranformed)
df2['Label'] = x.target.astype('string')
```

```python
df2.columns = ['PCA1', 'PCA2', 'Label']
```

```python
sns.scatterplot(data=df2, x="PCA1", y="PCA2", hue="Label")
plt.title('Visualization of MNIST using PCA')
plt.xlabel('PCA1')
plt.ylabel('PC2')
plt.show()
```

```python
sns.scatterplot(data=df, x="first_axis", y="second_axis", hue="Label")
plt.title('Visualization of MNIST using T-sne')
plt.xlabel('First axis')
plt.ylabel('second axis')
plt.show()
```
